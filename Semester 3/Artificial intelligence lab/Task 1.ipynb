```python
Tasks = []

def add_tasks():
    total_value = int(input("Enter the number of tasks you want to add: "))
    for i in range(total_value):
        task = input("Enter the task: ")
        priority = int(input("Enter the priority: "))
        Tasks.append({"task": task, "done": False, "priority": priority})
        print("Task added successfully.")

def show_tasks():
    if not Tasks:
        print("No tasks added till now.")
    else:
        for index, task in enumerate(Tasks, start=1):
            status = "Done" if task["done"] else "Not Done"
            print(f"{index}. {task['task']} - {status}")

def done_tasks():
    task_number = int(input("Enter task number: ")) - 1
    if 0 <= task_number < len(Tasks):
        Tasks[task_number]["done"] = True
        print("Task is done.")
    else:
        print("Invalid number.")

def main():
    while True:
        print("To Do Tasks: ")
        print("1. Add Tasks")
        print("2. Show Tasks")
        print("3. Done Tasks")
        print("4. Exit")
        
        choice = int(input("Enter your Choice: "))
        if choice == 1:
            add_tasks()
        elif choice == 2:
            show_tasks()
        elif choice == 3:
            done_tasks()
        elif choice == 4:
            print("Exiting To-Do Tasks")
            break
        else:
            print("Invalid Choice")

if __name__ == "__main__":
    main()
```
